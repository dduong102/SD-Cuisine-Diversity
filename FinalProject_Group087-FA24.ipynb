Next, we need to clean up this data set before we start wrangling it. For our question, the `age_group`, `sex`, and `yr_id` columns are irrelevant because we are only concerned with zip code, ethnicity, and population values in our scope of work. Therefore, we will remove those the `age_group` and `sex` columns, and combine the `population` column values for each ethnicity (because the total population for each ethnicity has been split apart into pieces by age groups and sexes). This way, only the variables of focus are displayed.

````python
sd_new = sd.groupby(['zip', 'yr_id', 'ethnicity']).sum().reset_index()
sd_new = sd_new.drop(columns=['age_group', 'sex', 'yr_id'])
sd_new.head(10)
sd_new.shape
````

Now, in preparation for our EDA, we will need to assign a `population_proportion` column, that will represent the population proportion based on the population count of that ethnicity type in a zip code, to the entire populatin count in that zip code.

For example:

1. Zip Code 91901 has 3203 hispanic counts.
2. The total population for 91901 is 17700. So the element for that row in the `population_proportion` column would be 3203/17700 ~= 0.18.
3. We repeat this for all of the other rows.

````python
sd_new['total_population'] = sd_new.groupby('zip')['population'].transform('sum')
sd_new['population_proportion'] = sd_new['population'] / sd_new['total_population']
sd_new['ethnicity'] = sd_new['ethnicity'].str.replace('Non-Hispanic, ', '', regex=False)
sd_new = sd_new.drop(columns=['total_population'])

sd_new = sd_new.rename(columns={
    'zip': 'Zipcode',
    'ethnicity': 'Ethnicity',
    'population': 'Population',
    'population_proportion': 'Population Proportion'
})
sd_new.head(10)
````

````python
# Mapping for cuisines to census ethnicity categories
# Used this Census Classification to closely match with SANDAG data: https://mcdc.missouri.edu/help/race-ethnicity.html#:~:text=On%20census%20surveys%2C%20an%20individual%20can%20report,person%20is%20of%20Hispanic%20origin%20or%20not.
cuisine_mapping = {
    'American': "White", 'Bagels': "White", 'Brewery': "White", 'Distillery': "White", 'Coffee': "White",
       'Japanese':"Asian", 'Cafe': "White", 'Mediterranean': "White", 'Lebanese':"White", 'Mexican': "Hispanic",
       'Salumeria': "White", 'Italian':"White", 'Vietnamese':"Asian", 'Breakfast':"White", 'Korean': "Asian",
       'Hot Dogs':"White", 'Pizza':"White", 'Dessert':"White", 'Winery': "white", 'French':"White", 'Cidery':"White",
       'Sandwiches':"White", 'Boba':"Asian", 'Market': "White", 'Creole/Cajun': "Black", 'Fusion': "Other",
       'Californian': "White", 'European': "White", 'Filipino':"Asian", 'Deli':"White", 'Thai': "Asian", 'Chinese': "Asian",
       'Sushi': "Asian", 'Indian': "Asian", 'Taiwanese': "Asian", 'Vegan': "Other", 'Seafood': "White", 'Argentinian':"Hispanic",
       'Healthy Stuff': "White", 'Doughnuts': "White", 'Bakery': "White", 'BBQ': "White", 'Hawaiian': "Hawaiian or Pacific Islander",
       'Steakhouse': "White", 'Meadery':"White", 'Caribbean': "Black", 'Greek': "White", 'Vegetarian': "Other",
       'Poke': "Hawaiian or Pacific Islander", 'Peruvian': "Hispanic", 'Shawarma': "White", 'Southern': "White", 'Middle-Eastern': "White",
       'Alcohol': "White", 'German': "White", 'International': "Other", 'Latin American': "Hispanic"
}

# Function to classify cuisines
def classify_cuisine(row):
    primary = cuisine_mapping.get(row["Cuisine"], "Other")
    secondary = cuisine_mapping.get(row["Cuisine.1"], "")
    if primary == secondary:
      return primary
    if primary == "Other" and secondary != "Other":
      return secondary
    if secondary == "Other" and primary != "Other":
      return primary
    return f"{primary}, {secondary}".strip(", ")

# Apply classification
filtered_df["Cuisine Census Ethnicity"] = filtered_df.apply(classify_cuisine, axis=1)
filtered_df.head(10)
print(filtered_df.shape)
````

Now with this information, we can finaly create a dataframe that has the following variables per zipcode: `Ethnicity`, `Restaurant Ethnicity Count`, and `Total Restaurant Ethnicities Combined`.

````python
# List of possible ethnicities
ethnicities = ['Hispanic', 'American Indian or Alaska Native', 'Asian', 'Black',
               'Hawaiian or Pacific Islander', 'Other', 'Two or More Races', 'White']

# Initialize an empty list to collect expanded rows
expanded_rows = []

# Loop through each row in the original DataFrame
for i, row in cuisine_to_ethnicities.iterrows():
    zipcode = row['Zipcode']
    census_ethnicities = row['Cuisine Census Ethnicity']

    # For each ethnicity in the list of all possible ethnicities
    for ethnicity in ethnicities:
        # Count how many times the ethnicity appears in the list for the current Zipcode
        ethnicity_count = census_ethnicities.count(ethnicity)

        # Add a new row for each ethnicity with the count
        expanded_rows.append({
            'Zipcode': zipcode,
            'Ethnicity': ethnicity,
            'Restaurant Ethnicity Count': ethnicity_count
        })

# Convert the expanded rows list to a DataFrame
expanded_df = pd.DataFrame(expanded_rows)
print(expanded_df)

zipcode_92111_df = expanded_df.loc[expanded_df['Zipcode'] == '92111']

# Print the filtered DataFrame
print(zipcode_92111_df)
````

````python
# Group by 'Zipcode' and sum the 'Ethnicity Count' for each Zipcode
total_ethnicities = expanded_df.groupby('Zipcode')['Restaurant Ethnicity Count'].sum().reset_index()

# Rename the column to 'Total Ethnicity Count'
total_ethnicities.rename(columns={'Ethnicity Count': 'Total Ethnicity Count'}, inplace=True)

restaurant_ethnicity_df = pd.merge(expanded_df, total_ethnicities, on='Zipcode', how='left')

restaurant_ethnicity_df = restaurant_ethnicity_df.rename(columns={
    'Restaurant Ethnicity Count_x': 'Restaurant Ethnicity Count',
    'Restaurant Ethnicity Count_y': 'Total Restaurant Ethnicities Combined',
    # Add more column name mappings as needed
})

restaurant_ethnicity_df.head(10)
````

### Section 2 of EDA - Comparing the similairty between Each Zipcode Distributions (Calculating Pearson's Coefficient)

First, we will need to pivot the dataframe in order to access the values easier for the next step.

Now that we have pivoted the dataframe, we can easily access two seprate arrays of each population proportion and restaurant ethnicity proportion, per zipcode. This will allow us to use the pearsonr() function, that takes in two arrays and calculates the final r corefficient form that.

For example: Zipcode 91902

Since the dsitribution of pearson's correlations are skewed, to find a better ultimate pearson correlation that encompass all of the neighborhoods into one correlation, we need to find the median. For demonstration purposes, we will show both the mean and median correlation, and show that the difference between those values can be very significant considering the following cutoffs:

Interval coefficient | Relationship level
---------------------|-------------------

These cuttoffs were made using the guide that Evans (1996) suggests for the absolute value of r:

````python
corr_list = list(correlations.values())

# Calculating the mean
mean = sum(corr_list) / len(corr_list)

# Sorting the values in ascending order
corr_list.sort()

# Calculating the median
n = len(corr_list)
if n % 2 == 1:
    # If odd number of values, median is the middle element
    median = corr_list[n // 2]
else:
    # If even number of values, median is the average of the two middle elements
    median = (corr_list[n // 2 - 1] + corr_list[n // 2]) / 2

print("Mean: " + str(mean) + ", Median: " + str(median))
````

### Section 3 of EDA - Chi Square Distribution

Although we have our median pearson's coefficient, we can conduct another statistical test to further support our results.

Therefore, we want to conduct a **chi-square test for independence** to determine whether the distribution of the ethnic population is independent (no relationship) of the distribution of restaurant ethnicities.

- `null hypothesis`: there is not a relationship between population demographics and restaurant offerings, and the two categories are independent
- `alternate hypothesis`: there is a relationship between population demographics and restaurant offerings

### Why is this important?

Having a low p-value (< 0.05, we will be using the standard alpha threshold for this test) displays a significant relationship between population demographics and restaurant offerings. This suggests that the availability of different types of restaurants may not be random but instead correlated with the ethnic makeup of the population.

```python
# This is formatted as code
from scipy.stats import chi2_contingency

# Example of running the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f'Chi2: {chi2}, p-value: {p}, Degrees of Freedom: {dof}')
```


In the context of this problem, the combination of the first ethnicity and the first zip code has a zero expected frequency. This suggests that for some areas, they do not have restaurants representing this specific ethnic group.

### Condusting the Chi-Square Test: Removing All-Zero Rows and Columns

Before performing the chi-square test, remove rows and columns with all-zero values to ensure valid and consistent results.

- **Avoid Errors**: Zero marginal totals lead to undefined expected frequencies.  
- **Consistency**: Including irrelevant categories skews the analysis.  
- **Accuracy**: Focuses the test on meaningful relationships.

````python
# Step 1: Create the contingency table
contingency_table = pd.crosstab(
    merged_df['Ethnicity'],
    merged_df['Zipcode'],
    values=merged_df['Restaurant Ethnicity Count'],
    aggfunc='sum'
).fillna(0)

# Step 2: Remove rows and columns with all-zero values
contingency_table = contingency_table.loc[(contingency_table != 0).any(axis=1)]
contingency_table = contingency_table.loc[:, (contingency_table != 0).any(axis=0)]

# Step 3: Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print results
print(f'Chi2: {chi2}, p-value: {p}, Degrees of Freedom: {dof}')
alpha = 0.05  # Significance level
if p < alpha:
    print("Reject the null hypothesis. There is a significant association between Ethnicity and Zipcode.")
else:
    print("Fail to reject the null hypothesis. No significant association between Ethnicity and Zipcode.")
````

The chi-square test results (χ² = 265.82, p = 0.0013, df = 200) rejected the null hypothesis. This indicate a **statistically significant relationship** between the ethnic demographics of San Diego neighborhoods and the types of ethnic cuisines offered by restaurants in those neighborhoods. This suggests that the availability of certain restaurant cuisines is likely influenced by the ethnic composition of the population.

Issues such as incomplete data coverage, particularly in neighborhoods like Mira Mesa and Scripps Ranch, the exclusion of dine-in establishments, and the broad ethnic categories in the census data all contribute to potential inaccuracies. If we were to further explore this question, I would start by looking deeper into the possibility that this relationship may not be as clean-cut as it appears to be on paper, since perhaps our data just appeared to be correlational simply because there wasn't enough data. As seen with zipcode 92111, maybe certain ethnic restaurant types dominate each neighborhood regardless of population demographics.
